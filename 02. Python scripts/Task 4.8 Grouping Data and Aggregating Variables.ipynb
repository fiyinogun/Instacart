### 4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
ords_prods_merged = ords_prods_merged.drop(columns = ['max_order'])
```

```python
ords_prods_merged = ords_prods_merged.drop(columns = ['loyalty_flag'])
```

```python
# Group data by user_id and transform number_of_items_ordered using max. Results stored in new column 'max_order'

ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['number_of_items_ordered'].transform(np.max)
```

```python
ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Frequency couny for loyalty flag

ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

###  6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

```python
# Group data by user_id and transform 'prices' using its averages by user. Results stored in new column 'avg_price'

ords_prods_merged['avg_price'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking results

ords_prods_merged[['user_id', 'loyalty_flag', 'avg_price']].head(100)
```

```python
# Low spender flag, stored in new column 'spending_habits'

ords_prods_merged.loc[ords_prods_merged['avg_price'] < 10, 'spending_habits'] = 'Low spender'
```

```python
# High spender flag, stored in new column 'spending_habits'

ords_prods_merged.loc[ords_prods_merged['avg_price'] >= 10, 'spending_habits'] = 'High spender'
```

```python
# Checking results

ords_prods_merged[['user_id', 'loyalty_flag', 'avg_price', 'spending_habits']].head(100)
```

###  7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

- If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
- If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
- If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Group data by user_id and transform 'days_since_prior_order' using its median by user. Results stored in new column 'median_days_last_order'

ords_prods_merged['median_days_last_order'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check results

ords_prods_merged[['user_id', 'median_days_last_order']].head(100)
```

```python
# Non-frequent customer flag, stored in new column 'customer_frequency'

ords_prods_merged.loc[ords_prods_merged['median_days_last_order'] > 20, 'customer_frequency'] = 'Non-frequent customer'
```

```python
# Regular customer flag, stored in 'customer_frequency'

ords_prods_merged.loc[(ords_prods_merged['median_days_last_order'] > 10) & (ords_prods_merged['median_days_last_order'] <= 20), 'customer_frequency'] = 'Regular customer'
```

```python
# Frequent customer flag, stored in 'customer_frequency'

ords_prods_merged.loc[ords_prods_merged['median_days_last_order'] <= 10, 'customer_frequency'] = 'Frequent customer'
```

```python
ords_prods_merged['customer_frequency'].value_counts(dropna = False)
```

```python
# Check results

ords_prods_merged[['user_id', 'median_days_last_order', 'customer_frequency']].head(100)
```
